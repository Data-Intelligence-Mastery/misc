```
#%% import lib and load data
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation as LDA

# !pip install flask-ngrok
# !pip install pyLDAvis

# from google.colab import drive
# drive.mount('/content/drive')

# load the updated news_table
df = pd.read_csv('drive/MyDrive/Colab Notebooks/Demo Projects/news_table.csv', )
df.sample(5)
```

```
df.dropna(inplace=True)

kwargs_vector = ({'max_df': 0.7},
                 {'min_df': 0.08},
                 {'ngram_range': (1, 2)})
vectorizer = CountVectorizer(*kwargs_vector, stop_words='english')
dtm = vectorizer.fit_transform(df['paragraph'])

kwargs_model = {'n_components': 8}     
model = LDA(**kwargs_model, random_state=42)
model.fit(dtm)

print('Completed training')
```

```
#%% pyLDAvis visualization

import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()

# pyLDAvis.sklearn.prepare(model, dtm, vectorizer)
vis_data = pyLDAvis.sklearn.prepare(model, dtm, vectorizer)
pyLDAvis.display(vis_data)
```

```
from flask_ngrok import run_with_ngrok
from flask import Flask
app = Flask(__name__)
run_with_ngrok(app)   #starts ngrok when the app is run
@app.route("/")
def home():
    return pyLDAvis.display(vis_data)

    # return "<h1>Running Flask on Google Colab!</h1>"
  
app.run()
```
